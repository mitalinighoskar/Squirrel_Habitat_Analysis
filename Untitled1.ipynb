```python
import matplotlib.pyplot as plt

data['Date'] = pd.to_datetime(data['Date'])

feeding_counts = data.groupby('Date')['Eating'].sum()

plt.figure(figsize=(12, 6))
feeding_counts.plot(kind='line')
plt.title('Squirrel Feeding Activity Over Time')
plt.xlabel('Date')
plt.ylabel('Feeding Activity Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

total_squirrels = len(data)
hand_feeding_count = data['Running'].sum()
non_hand_feeding_count = total_squirrels - hand_feeding_count

percentages = [hand_feeding_count, non_hand_feeding_count]

labels = ['Hand Feeding', 'Not Hand Feeding']

plt.figure(figsize=(6, 6))
plt.pie(percentages, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Squirrels that Hand Feed')
plt.show()
```

```python
# Stacked bar chart for activities by primary fur color
activity_columns = ['Running', 'Chasing', 'Climbing', 'Eating']
fur_color_grouped = data.groupby('Primary Fur Color')[activity_columns].sum()
fur_color_grouped.plot(kind='bar', stacked=True)
plt.xlabel('Primary Fur Color')
plt.ylabel('Count')
plt.title('Squirrel Activities by Primary Fur Color')
plt.show()
```

```python
# Bar chart for squirrel activities by shift
shift_grouped = data.groupby('Shift')[activity_columns].sum()
shift_grouped.plot(kind='bar')
plt.xlabel('Shift')
plt.ylabel('Count')
plt.title('Squirrel Activities by Shift')
plt.show()
```

```python
# Pie chart for running and chasing activities
running_chasing_counts = data[['Running', 'Chasing']].sum()
labels = ['Running', 'Chasing']
plt.pie(running_chasing_counts, labels=labels, autopct='%1.1f%%')
plt.title('Proportion of Running and Chasing Activities')
plt.show()
```
